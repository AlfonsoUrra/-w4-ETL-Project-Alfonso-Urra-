# Descripción del proyecto

Este proyecto es un ETl, por sus siglas en inglés. Extract, transform and load. La temática de los datos gira entorno al cambio climático. He buscado y trabajado con diferentes archivos quer pueden explicar y relacionar diferentes tópicos sobre este tema.

# Características de los datos

He utilizado 3 archivos de 2 orígenes diferentes, archivo csv de Kaggle, api de World bank open data y un scrappeo de 'https://cloford.com/resources/codes/index.htm

Los datos aportan información sobre las diferentes emisiones de Co2 relacionadas sobretodo con el GDP y también aportan para el mismo periodo la inversión que ha realizado cada país respecto a su GDP en Sanidad y salud.

El último scrappeo recoge una lista de países, continentes y sus iso para una mayor fluidez a la hora de estudiar los datos y poder relacionarlos de forma más limpia.


# Objetivo del proyecto


Estudiando los datos y con perspectiva de continuar este proyecto, mi objetivo es claro: estudiar los diferentes datos que pueden ayudar a explicar el cambio climático y sus consecuencias. Dependiendo de los datos que pueda ir encontrando en el futuro , orientaré mi estudio hacia un lado u otro. 

En este punto, y con los archivos seleccionados hasta hoy, mis perspectiva principal es analizar para cada país y continente, el número y tipo de emisiones totales, principalmente en los últimos 30 años y como estas están relacionadas con el incremento en gasto en sanidad y la pérdida de masa forestal.

Dado que el proyecto todavía está en fase de obtención de datos y todavía no tengo definido el camino a seguir, he preferido dejar las conclusiones para más adelante y centrarme en organizar los datos actuales de tal forma que pueda en un futuro obtener conclusiones de forma eficaz.

He mantenido el data orginal lo más posible para no alterar futuros estudios, haciendo mas trabajo de organización que de limpieza.

# Proceso

Tenía una idea preconcebida de que hacer y como hacerlo, pero al encontrarme  con los datos tuve que redireccionar mi estudio y lo centré en varios puntos:

    1. Análisis por país y continente y año.
    
    2. Evolución de las emisiones
    
    3. Evoluciión del GDP y la población
    
    4. Evolución del gasto en Health.
    
    5. Volúmen masa forestal.
    
    
Sobre esta base tuve mucha más facilidad a la hora de organizar mis tablas. 

Eliminé datos que no me interesaban, organicé los archivos para darles homogeneidad, y extraje datos que estaban escritos de forma poca legible. Así fuese más fácil interpretarlos y relacionarlos.

Al tener las 3 tablas ya organizadas y limpias, pude concatenarlas de forma eficiente. Acumulé bajo la columna 'Indicator' todos los datos relevantes sobre los que quería organizar mis archivos, y los relacioné a través de los países y sus 'ISO'. Teniendo una evolución histórica de cada país sobre los siguientes temas : 

       1. 'Capital health expenditure (% of GDP)',
       
       2. 'Current health expenditure (% of GDP)',
       
       3. 'Domestic general government health expenditure (% of GDP)',
       
       4. 'Public spending on education, total (% of GDP)', 'co2_per_gdp',
       
       5. 'consumption_co2_per_gdp', 'gdp', 'energy_per_gdp',
       
       6. 'Forest area (% of land area)'

Después solo quedaba la parte de subirlos a SQL y comprobar que todo funcionaba a través de una query.
